```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

# Create a grid of x and y values
x = np.linspace(-2, 3, 400)
y = np.linspace(-3, 1, 400)
X, Y = np.meshgrid(x, y)

# Compute f(x, y)
Z = X**3 + Y**3 - 2*X**2 + 3*Y**2 - 8

# 3D Surface Plot with Rotation
fig = plt.figure(figsize=(12, 6))

# First subplot: 3D surface plot
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
surface = ax1.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax1.set_title('3D Surface Plot')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
# rotate the plot
# ax1.view_init(elev=30, azim=90)
fig.colorbar(surface, ax=ax1, shrink=0.5, aspect=10)

# Enable interactive rotation with the mouse
plt.ion()

# Function to update the angle of view for animation
def rotate(angle):
    ax1.view_init(elev=120, azim=angle)
    return ax1

# Create an animation rotating the plot from 0 to 360 degrees
ani = FuncAnimation(fig, rotate, frames=np.arange(0, 360, 2), interval=50, blit=False)

# Second subplot: 2D contour plot
ax2 = fig.add_subplot(1, 2, 2)
contour = ax2.contourf(X, Y, Z, levels=50, cmap='viridis')
ax2.set_title('2D Contour Plot')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

fig.colorbar(contour, ax=ax2)

plt.tight_layout()
plt.show()
```
